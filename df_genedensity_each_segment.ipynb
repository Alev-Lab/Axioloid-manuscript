```python
def dfmaker(tissue_number):

    genelist = all_df["Gene"].tolist()

    roi_index = roi_df.position.tolist()
    roi_area = roi_df.area.tolist()
    roi_x = roi_df.mass_x.tolist()
    roi_y = roi_df.mass_y.tolist()

    roi_list = []
    for i in range(1, len(roi_index)+1):
        a= f"r{i}"
        roi_list.append(a)
    
        
    df1 = pd.DataFrame(columns = roi_list) #for number of amplicon
    df2 = pd.DataFrame(columns = roi_list) #for density of amplicon (1000 um2)

    for i in range(len(genelist)):
        genename = genelist[i]
        x = rna_position_df.loc[rna_position_df.Gene == genename, "x_pos"].tolist()
        y = rna_position_df.loc[rna_position_df.Gene == genename, "y_pos"].tolist()
        pos_gene = list(zip(x,y))

        num_amp = []
        density_amp = []

        for j in range(len(roi_index)):
            roi=np.array(list(zip(*np.where(labels==roi_index[j]))))
            pos_x, pos_y= roi[:,1].tolist(), roi[:,0].tolist()
            roi_pos = list(zip(pos_x, pos_y))

            p = set(roi_pos) & set(pos_gene)
            num_amp.append(len(p))

            area = roi_area[j]
            num_amp_density = len(p)*39062.5/area   # density in 1000 um2...1px = 0.16 um x 0.16 um
            density_amp.append(num_amp_density)




        df1.loc[genename] = num_amp
        df2.loc[genename] = density_amp

    mesp2_list = df1.loc["MESP2"].tolist()
    idx_mesp2max = mesp2_list.index(max(mesp2_list))

    distance_pxl1 = []
    for i in range(len(roi_x)-1):
        d2 = (roi_x[i]-roi_x[i+1])**2 + (roi_y[i]-roi_y[i+1])**2
        d = math.sqrt(d2)
        distance_pxl1.append(d)
    distance_pxl1.append(0)

    distance_pxl2 = []
    distance_microm = []
    a = sum(distance_pxl1[idx_mesp2max:len(roi_x)])
    for i in range(len(roi_x)):
        d3 = sum(distance_pxl1[i:len(roi_x)]) - a
        distance_pxl2.append(d3)
        d4 = d3*0.16
        distance_microm.append(d4)


    df1.loc["distance_pxl"] = distance_pxl2
    df2.loc["distance_pxl"] = distance_pxl2

    df1.loc["distance_micro-m"] = distance_microm
    df2.loc["distance_micro-m"] = distance_microm



    df1.to_csv(f"{inputdir}\\output\\{tissue_number}_amplicon_number_roi.csv")
    df2.to_csv(f"{inputdir}\\output\\{tissue_number}_amplicon_density_roi.csv")

```
